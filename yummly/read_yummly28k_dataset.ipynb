```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from IPython.display import Image, display

import pandas as pd
import numpy as np
import json
%matplotlib notebook
```

```python
cols = ['id_meta', 'name', 'cuisine', 'ingredients']
df = pd.DataFrame(columns=cols)

DATASET_SIZE = 27638

for i in range(DATASET_SIZE):
    filename = '/projects/kawa7978/metadata27638/meta%05d.json'%(i+1)
    f = open(filename, 'r')
    d = json.load(f)
    new_d = ({'id_meta': d['id'], 'name': d['name'], 'cuisine': d['attributes']['cuisine'], 'ingredients': d['ingredientLines']})
    df = df.append(new_d, ignore_index=True)
```
